```python
# import pandas
import numpy as np
import pandas as pd
# import matplotlin
import matplotlib.pyplot as plt
# import seaborn
import seaborn as sns
%matplotlib inline

def create_boxplot(df, i, ts, cs):

    data5500itters = df[df.Itters == i] # Set the number of itteration
    data5500itters = data5500itters[data5500itters.Type != 'random2']
    data5500i30000p = data5500itters[data5500itters.Total_Size == ts]# Keep only those with the total size 
    data5500i30000p = data5500itters[data5500itters.Current_Size == cs]# Keep only those with the total size 
    
    return sns.boxplot(y='Area', x='Type', data=data5500i30000p, width=0.5,palette="colorblind").set_title("Comparison of Area of Madelbrot Set\n s = {} and i = {} ".format(cs,i))
```
